```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import matplotlib.colors as mcolors
import os

# Set publication-quality parameters with VERY LARGE FONTS
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman', 'DejaVu Serif']
plt.rcParams['font.size'] = 24
plt.rcParams['axes.labelsize'] = 28
plt.rcParams['axes.titlesize'] = 32
plt.rcParams['xtick.labelsize'] = 22
plt.rcParams['ytick.labelsize'] = 22
plt.rcParams['legend.fontsize'] = 22
plt.rcParams['figure.titlesize'] = 36

# Define color scheme - professional blues and grays
color_1d = '#3498db'  # Blue
color_2d = '#2ecc71'  # Green  
color_axis = '#34495e'  # Dark gray
color_text = '#2c3e50'  # Darker gray

# ============================================================================
# FIGURE 1: 1D and 2D Arrays
# ============================================================================
fig1 = plt.figure(figsize=(18, 8))

# ============================================================================
# SUBPLOT 1: 1D Array (Vector)
# ============================================================================
ax1 = plt.subplot(1, 2, 1)
ax1.set_xlim(-0.5, 5.5)
ax1.set_ylim(-1.2, 3.8)
ax1.axis('off')
ax1.set_aspect('equal')

# Title - VERY BIG
ax1.text(2.5, 3.3, '1D Array (Vector)', fontsize=36, weight='bold', 
         ha='center', color=color_text)

# Draw 1D array boxes
array_1d = [12, 15, 18, 21, 24]
box_width = 0.85
box_height = 0.85
start_x = 0.5

for i, val in enumerate(array_1d):
    # Draw box
    rect = FancyBboxPatch((start_x + i, 0.5), box_width, box_height,
                          boxstyle="round,pad=0.05", 
                          edgecolor=color_1d, facecolor=color_1d, 
                          alpha=0.3, linewidth=4)
    ax1.add_patch(rect)
    
    # Add value - VERY BIG FONT
    ax1.text(start_x + i + box_width/2, 0.5 + box_height/2, str(val),
            ha='center', va='center', fontsize=26, weight='bold',
            color=color_text)
    
    # Add index below - BIG FONT
    ax1.text(start_x + i + box_width/2, -0.05, f'[{i}]',
            ha='center', va='center', fontsize=22, color=color_axis,
            weight='bold')

# Draw axis 0 arrow - VERY BIG ARROW
arrow = FancyArrowPatch((0.2, 2.1), (5.2, 2.1),
                       arrowstyle='->', mutation_scale=45, 
                       linewidth=5, color=color_axis)
ax1.add_patch(arrow)
ax1.text(2.7, 2.6, 'axis 0', ha='center', fontsize=28, 
         weight='bold', color=color_axis)

# Add shape information - BIG FONT
ax1.text(2.5, -0.8, 'Shape: (5,)', ha='center', fontsize=24,
         style='italic', color=color_text, weight='bold')

# ============================================================================
# SUBPLOT 2: 2D Array (Matrix)
# ============================================================================
ax2 = plt.subplot(1, 2, 2)
ax2.set_xlim(-2.2, 5.5)
ax2.set_ylim(-1.5, 5.2)
ax2.axis('off')
ax2.set_aspect('equal')

# Title - VERY BIG
ax2.text(1.8, 4.6, '2D Array (Matrix)', fontsize=36, weight='bold',
         ha='center', color=color_text)

# Draw 2D array
array_2d = np.array([[12, 15, 18],
                     [21, 24, 27],
                     [30, 33, 36]])
rows, cols = array_2d.shape
box_size = 0.85
start_x_2d = 0.2
start_y_2d = 0.2

for i in range(rows):
    for j in range(cols):
        # Draw box
        rect = FancyBboxPatch((start_x_2d + j*box_size, start_y_2d + (rows-1-i)*box_size),
                             box_size, box_size,
                             boxstyle="round,pad=0.03",
                             edgecolor=color_2d, facecolor=color_2d,
                             alpha=0.3, linewidth=4)
        ax2.add_patch(rect)
        
        # Add value - VERY BIG FONT
        ax2.text(start_x_2d + j*box_size + box_size/2,
                start_y_2d + (rows-1-i)*box_size + box_size/2,
                str(array_2d[i, j]),
                ha='center', va='center', fontsize=24, weight='bold',
                color=color_text)

# Row indices (axis 0) - BIG FONT
for i in range(rows):
    ax2.text(-0.25, start_y_2d + (rows-1-i)*box_size + box_size/2,
            f'[{i}]', ha='right', va='center', fontsize=22, color=color_axis,
            weight='bold')

# Column indices (axis 1) - BIG FONT
for j in range(cols):
    ax2.text(start_x_2d + j*box_size + box_size/2, -0.15,
            f'[{j}]', ha='center', va='center', fontsize=22, color=color_axis,
            weight='bold')

# Axis 0 arrow (vertical) - VERY BIG ARROW
y_start = start_y_2d + rows*box_size + 0.5
y_end = start_y_2d - 0.5
arrow_0 = FancyArrowPatch((-0.9, y_start), (-0.9, y_end),
                         arrowstyle='->', mutation_scale=45,
                         linewidth=5, color=color_axis)
ax2.add_patch(arrow_0)
ax2.text(-1.5, (y_start + y_end)/2, 'axis 0', rotation=90,
         ha='center', va='center', fontsize=28, weight='bold',
         color=color_axis)

# Axis 1 arrow (horizontal) - VERY BIG ARROW
x_start = start_x_2d
x_end = start_x_2d + cols*box_size
arrow_1 = FancyArrowPatch((x_start - 0.4, -0.75), (x_end + 0.4, -0.75),
                         arrowstyle='->', mutation_scale=45,
                         linewidth=5, color=color_axis)
ax2.add_patch(arrow_1)
ax2.text((x_start + x_end)/2, -1.2, 'axis 1', ha='center',
         fontsize=28, weight='bold', color=color_axis)

# Add shape information - BIG FONT
ax2.text(1.8, -0.5, 'Shape: (3, 3)', ha='center', fontsize=24,
         style='italic', color=color_text, weight='bold')

plt.tight_layout()

# Save Figure 1
save_dir = r'C:\Users\ktripat\Dropbox\book\Figures\Create Figures\chapter-2'
os.makedirs(save_dir, exist_ok=True)
os.makedirs('/tmp', exist_ok=True)

for directory in [save_dir, '/tmp']:
    try:
        plt.savefig(os.path.join(directory, 'numpy_1d_2d_arrays.png'), 
                    dpi=300, bbox_inches='tight', facecolor='white', edgecolor='none')
        plt.savefig(os.path.join(directory, 'numpy_1d_2d_arrays.pdf'), 
                    bbox_inches='tight', facecolor='white', edgecolor='none')
        print(f"Figure 1 (1D/2D) saved to: {directory}")
    except Exception as e:
        print(f"Could not save Figure 1 to {directory}: {e}")

plt.show()

# ============================================================================
# FIGURE 2: 3D Array (Tensor) - LARGE AND CLEAR
# ============================================================================

# Create custom 3D arrow class
class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        super().__init__((0, 0), (0, 0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def do_3d_projection(self, renderer=None):
        from mpl_toolkits.mplot3d import proj3d
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, self.axes.M)
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))
        return np.min(zs)

fig2 = plt.figure(figsize=(16, 12))
ax3 = fig2.add_subplot(111, projection='3d')

ax3.set_xlim(-1.5, 5)
ax3.set_ylim(-1.5, 6)
ax3.set_zlim(-1.5, 5)
ax3.view_init(elev=22, azim=35)
ax3.grid(False)
ax3.set_xticks([])
ax3.set_yticks([])
ax3.set_zticks([])
ax3.xaxis.pane.fill = False
ax3.yaxis.pane.fill = False
ax3.zaxis.pane.fill = False
ax3.xaxis.pane.set_edgecolor('none')
ax3.yaxis.pane.set_edgecolor('none')
ax3.zaxis.pane.set_edgecolor('none')

# Title - HUGE
ax3.text2D(0.5, 0.97, '3D Array (Tensor)', fontsize=48, weight='bold',
           ha='center', transform=ax3.transAxes, color=color_text)

# Draw 3D tensor as stacked layers - Shape (3, 4, 3)
depth_layers = 3
n_rows = 4
n_cols = 3
cube_size = 0.85
spacing = 0.3

# Create color gradient for depth layers: red -> purple -> blue
depth_colors = ['#e74c3c', '#9b59b6', '#3498db']  # Red -> Purple -> Blue

# Draw layers with different colors
for d in range(depth_layers):
    z_offset = d * (cube_size + spacing)
    layer_color = depth_colors[d]
    
    # Higher alpha for front layers
    alpha_val = 0.3 + 0.15 * (depth_layers - d) / depth_layers
    
    for h in range(n_rows):
        for w in range(n_cols):
            # Define cube vertices
            x = w * cube_size
            y = h * cube_size
            z = z_offset
            
            # Draw all cubes
            vertices = [
                [x, y, z],
                [x + cube_size, y, z],
                [x + cube_size, y + cube_size, z],
                [x, y + cube_size, z],
                [x, y, z + cube_size],
                [x + cube_size, y, z + cube_size],
                [x + cube_size, y + cube_size, z + cube_size],
                [x, y + cube_size, z + cube_size]
            ]
            
            # Define cube faces
            faces = [
                [vertices[0], vertices[1], vertices[5], vertices[4]],  # Bottom
                [vertices[2], vertices[3], vertices[7], vertices[6]],  # Top
                [vertices[0], vertices[3], vertices[7], vertices[4]],  # Left
                [vertices[1], vertices[2], vertices[6], vertices[5]],  # Right
                [vertices[0], vertices[1], vertices[2], vertices[3]],  # Front
                [vertices[4], vertices[5], vertices[6], vertices[7]]   # Back
            ]
            
            # Draw cube
            cube = Poly3DCollection(faces, facecolors=layer_color, 
                                   linewidths=2, edgecolors=layer_color,
                                   alpha=alpha_val)
            ax3.add_collection3d(cube)

# Add axis labels AT THE FRONT with HUGE text and arrows
# Axis 2 (width/columns) - FRONT, horizontal
ax3.text(1.3, -1.0, -0.8, 'axis 2', fontsize=36, weight='bold',
         color=color_axis, ha='center')
arrow2 = Arrow3D([-0.5, 2.8], [-0.7, -0.7], [-0.5, -0.5],
                mutation_scale=50, lw=6, arrowstyle='->', color=color_axis)
ax3.add_artist(arrow2)

# Axis 1 (height/rows) - FRONT, vertical
ax3.text(-1.0, 1.7, -0.8, 'axis 1', fontsize=36, weight='bold',
         color=color_axis, ha='center')
arrow1 = Arrow3D([-0.7, -0.7], [-0.5, 3.6], [-0.5, -0.5],
                mutation_scale=50, lw=6, arrowstyle='->', color=color_axis)
ax3.add_artist(arrow1)

# Axis 0 (depth) - showing the stacking direction
ax3.text(3.2, -1.0, 2.0, 'axis 0', fontsize=36, weight='bold',
         color=color_axis, ha='center')
arrow0 = Arrow3D([2.8, 2.8], [-0.7, -0.7], [-0.5, 3.0],
                mutation_scale=50, lw=6, arrowstyle='->', color=color_axis)
ax3.add_artist(arrow0)

# Add shape information - HUGE FONT
ax3.text2D(0.5, 0.04, 'Shape: (3, 4, 3)', fontsize=32, style='italic',
           ha='center', transform=ax3.transAxes, color=color_text, weight='bold')

# Add color legend for depth layers - BIGGER
legend_y = 0.88
ax3.text2D(0.5, legend_y, 'Depth Layers:', fontsize=28, weight='bold',
           ha='center', transform=ax3.transAxes, color=color_text)
depth_labels = ['Front (Red)', 'Middle (Purple)', 'Back (Blue)']
for i, (color, label) in enumerate(zip(depth_colors, depth_labels)):
    ax3.text2D(0.5, legend_y - 0.055*(i+1), label, fontsize=24,
               ha='center', transform=ax3.transAxes, color=color,
               weight='bold', bbox=dict(boxstyle='round,pad=0.4', 
               facecolor=color, alpha=0.35, edgecolor=color, linewidth=3))

plt.tight_layout()

# Save Figure 2
for directory in [save_dir, '/tmp']:
    try:
        plt.savefig(os.path.join(directory, 'numpy_3d_tensor.png'), 
                    dpi=300, bbox_inches='tight', facecolor='white', edgecolor='none')
        plt.savefig(os.path.join(directory, 'numpy_3d_tensor.pdf'), 
                    bbox_inches='tight', facecolor='white', edgecolor='none')
        print(f"Figure 2 (3D Tensor) saved to: {directory}")
    except Exception as e:
        print(f"Could not save Figure 2 to {directory}: {e}")


plt.show()
```
